Задание 1
Пользователь вводит с клавиатуры набор чисел. Полученные числа необходимо сохранить в список (тип
списка нужно выбрать в зависимости от поставленной
ниже задачи)

```python
class NumberManager:
    def __init__(self):
        self._numbers = []

    def add_number(self, num):
        self._numbers.append(num)

    def remove_number(self, num):
        if num in self._numbers:
            self._numbers.remove(num)

    def get_numbers(self):
        return self._numbers

    def get_min(self):
        return min(self._numbers) if self._numbers else None

    def get_max(self):
        return max(self._numbers) if self._numbers else None

    def get_sum(self):
        return sum(self._numbers) if self._numbers else 0

    def is_empty(self):
        return len(self._numbers) == 0


class NumberManagerApp:
    def __init__(self):
        self.manager = NumberManager()

    def run(self):
        while True:
            print("\n1 - Добавить число")
            print("2 - Удалить число")
            print("3 - Показать список")
            print("4 - Показать минимальное число")
            print("5 - Показать максимальное число")
            print("6 - Показать сумму чисел")
            print("7 - Выход")
            choice = input("Выберите действие: ")
            if choice == "1":
                num = input("Введите число: ")
                if num.isdigit() or (num.startswith('-') and num[1:].isdigit()):
                    self.manager.add_number(int(num))
                else:
                    print("Некорректный ввод")
            elif choice == "2":
                if not self.manager.is_empty():
                    num_to_remove = input("Введите число для удаления: ")
                    if num_to_remove.isdigit() or (num_to_remove.startswith('-') and num_to_remove[1:].isdigit()):
                        self.manager.remove_number(int(num_to_remove))
                    else:
                        print("Некорректный ввод")
                else:
                    print("Список пуст")
            elif choice == "3":
                numbers = self.manager.get_numbers()
                print("Список:", numbers if numbers else "пуст")
            elif choice == "4":
                mn = self.manager.get_min()
                print("Минимальное число:", mn if mn is not None else "список пуст")
            elif choice == "5":
                mx = self.manager.get_max()
                print("Максимальное число:", mx if mx is not None else "список пуст")
            elif choice == "6":
                s = self.manager.get_sum()
                print("Сумма чисел:", s if not self.manager.is_empty() else "список пуст")
            elif choice == "7":
                break
            else:
                print("Некорректный выбор")


if __name__ == "__main__":
    app = NumberManagerApp()
    app.run()
```

Реализуйте класс стека для работы со строками (стек
строк).
Стек должен иметь фиксированный размер. 

```python
class FixedStringStack:
    def __init__(self, size):
        self._stack = []
        self._size = size

    def push(self, item):
        if not self.is_full():
            self._stack.append(item)

    def pop(self):
        if not self.is_empty():
            return self._stack.pop()

    def top(self):
        if not self.is_empty():
            return self._stack[-1]

    def is_empty(self):
        return len(self._stack) == 0

    def is_full(self):
        return len(self._stack) == self._size


class FixedStackApp:
    def __init__(self):
        self.stack = None

    def run(self):
        size_input = input("Введите размер стека: ")
        if size_input.isdigit():
            self.stack = FixedStringStack(int(size_input))
        else:
            print("Некорректный ввод")
            return

        while True:
            print("\n1 - Добавить строку в стек")
            print("2 - Извлечь строку из стека")
            print("3 - Проверить, пуст ли стек")
            print("4 - Проверить, полон ли стек")
            print("5 - Показать верхний элемент")
            print("6 - Выход")
            choice = input("Выберите действие: ")
            if choice == "1":
                item = input("Введите строку: ")
                self.stack.push(item)
            elif choice == "2":
                popped = self.stack.pop()
                if popped is not None:
                    print("Извлечено:", popped)
                else:
                    print("Стек пуст")
            elif choice == "3":
                print("Стек пуст" if self.stack.is_empty() else "Стек не пуст")
            elif choice == "4":
                print("Стек полон" if self.stack.is_full() else "Стек не полон")
            elif choice == "5":
                top_item = self.stack.top()
                if top_item is not None:
                    print("Верхний элемент:", top_item)
                else:
                    print("Стек пуст")
            elif choice == "6":
                break
            else:
                print("Некорректный выбор")


if __name__ == "__main__":
    app = FixedStackApp()
    app.run()
```
