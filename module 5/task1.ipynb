Написать программу «справочник».Создать два списка
целых. Один список хранит идентификационные коды,
второй — телефонные номера. Реализовать меню для
пользователя:
■ Отсортировать по идентификационным кодам;
■ Отсортировать по номерам телефона;
■ Вывести список пользователей с кодами и телефонами;
■ Выход.

```python
# Инициализация списков
ids = [101, 203, 102, 305, 104]
phones = [5551234, 5555678, 5552345, 5556789, 5553456]

# Функция для вывода списка пользователей
def show_users(ids, phones):
    print("\nСписок пользователей:")
    for user_id, phone in zip(ids, phones):
        print(f"ID: {user_id}, Телефон: {phone}")
    print()

# Функция для сортировки по идентификационным кодам
def sort_by_ids(ids, phones):
    sorted_pairs = sorted(zip(ids, phones))
    return [pair[0] for pair in sorted_pairs], [pair[1] for pair in sorted_pairs]

# Функция для сортировки по номерам телефона
def sort_by_phones(ids, phones):
    sorted_pairs = sorted(zip(phones, ids))
    return [pair[1] for pair in sorted_pairs], [pair[0] for pair in sorted_pairs]

# Главное меню
def main():
    global ids, phones
    while True:
        print("Меню:")
        print("1. Отсортировать по идентификационным кодам")
        print("2. Отсортировать по номерам телефона")
        print("3. Вывести список пользователей")
        print("4. Выход")
        choice = input("Введите номер операции: ")

        if choice == "1":
            ids, phones = sort_by_ids(ids, phones)
            print("\nСписок отсортирован по идентификационным кодам.\n")
        elif choice == "2":
            ids, phones = sort_by_phones(ids, phones)
            print("\nСписок отсортирован по номерам телефона.\n")
        elif choice == "3":
            show_users(ids, phones)
        elif choice == "4":
            print("Выход из программы.")
            break
        else:
            print("Некорректный ввод. Попробуйте снова.")

if __name__ == "__main__":
    main()
```
