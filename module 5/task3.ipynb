Есть четыре списка целых. Необходимо их объединить
в пятом списке. Полученный результат в зависимости от
выбора пользователя отсортировать по убыванию или
возрастанию. Найти значение, введенное пользователем,
с использованием линейного поиска.

```python
# Инициализация списков
list1 = [3, 7, 2]
list2 = [10, 5, 8]
list3 = [1, 4, 6]
list4 = [9, 11, 12]

# Объединение списков в один
combined_list = list1 + list2 + list3 + list4

# Функция для линейного поиска значения
def linear_search(lst, target):
    for i, value in enumerate(lst):
        if value == target:
            return i  # Возвращает индекс найденного элемента
    return -1  # Если элемент не найден

# Главное меню
def main():
    global combined_list
    while True:
        print("\nМеню:")
        print("1. Отсортировать по возрастанию")
        print("2. Отсортировать по убыванию")
        print("3. Найти значение в списке")
        print("4. Выход")
        choice = input("Введите номер операции: ")

        if choice == "1":
            combined_list.sort()
            print("\nСписок отсортирован по возрастанию:")
            print(combined_list)
        elif choice == "2":
            combined_list.sort(reverse=True)
            print("\nСписок отсортирован по убыванию:")
            print(combined_list)
        elif choice == "3":
            target = int(input("Введите значение для поиска: "))
            index = linear_search(combined_list, target)
            if index != -1:
                print(f"Значение {target} найдено на позиции {index}.")
            else:
                print(f"Значение {target} не найдено.")
        elif choice == "4":
            print("Выход из программы.")
            break
        else:
            print("Некорректный ввод. Попробуйте снова.")

if __name__ == "__main__":
    main()
```
