Есть четыре списка целых. Необходимо объединить
в пятом списке только те элементы, которые уникальны
для каждого списка. Полученный результат в зависимости
от выбора пользователя отсортировать по убыванию или
возрастанию. Найти значение, введенное пользователем,
с использованием бинарного поиска.

```python
# Инициализация списков
list1 = [3, 7, 2, 8]
list2 = [10, 5, 8, 7]
list3 = [1, 4, 6, 10]
list4 = [9, 11, 12, 4]

# Функция для получения уникальных элементов из списка
def get_unique_elements(*lists):
    all_elements = [item for lst in lists for item in lst]
    return [item for item in all_elements if all_elements.count(item) == 1]

# Функция для бинарного поиска значения
def binary_search(lst, target):
    left, right = 0, len(lst) - 1
    while left <= right:
        mid = (left + right) // 2
        if lst[mid] == target:
            return mid  # Возвращает индекс найденного элемента
        elif lst[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1  # Если элемент не найден

# Главное меню
def main():
    # Получение уникальных элементов из всех списков
    combined_list = get_unique_elements(list1, list2, list3, list4)

    while True:
        print("\nМеню:")
        print("1. Отсортировать по возрастанию")
        print("2. Отсортировать по убыванию")
        print("3. Найти значение в списке")
        print("4. Выход")
        choice = input("Введите номер операции: ")

        if choice == "1":
            combined_list.sort()
            print("\nСписок отсортирован по возрастанию:")
            print(combined_list)
        elif choice == "2":
            combined_list.sort(reverse=True)
            print("\nСписок отсортирован по убыванию:")
            print(combined_list)
        elif choice == "3":
            if not combined_list:
                print("Список пуст. Добавьте уникальные элементы.")
                continue
            target = int(input("Введите значение для поиска: "))
            combined_list.sort()  # Для бинарного поиска список должен быть отсортирован
            index = binary_search(combined_list, target)
            if index != -1:
                print(f"Значение {target} найдено на позиции {index}.")
            else:
                print(f"Значение {target} не найдено.")
        elif choice == "4":
            print("Выход из программы.")
            break
        else:
            print("Некорректный ввод. Попробуйте снова.")

if __name__ == "__main__":
    main()
```
